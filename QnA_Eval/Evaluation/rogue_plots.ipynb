```python
import json
import matplotlib.pyplot as plt

rouge_scores = json.load(open('rouge_scores.json'))
```

```python
models = ["open_source", "tf-idf", "bm25", "vision", "BOW", "tf_idf_bm25_open"]

ratios = ["1_0", "1_1", "1_2"]

score = {}

for model in models:
    for ratio in ratios:
        file_name = f"{model}_{ratio}_top_100_response.json"
        file_name_2 = f"{model}_{ratio}"
        for rouge_score in rouge_scores:
            if rouge_score['file_name'] == file_name:
                score[file_name_2] = rouge_score['average_rouge_l']
```

```python
score
```

```python
# Extract keys and values for plotting
labels = list(score.keys())
values = list(score.values())

# Create the plot
plt.figure(figsize=(12, 6))
plt.barh(labels, values, color='skyblue', edgecolor='black')

# Set fixed axis limits for consistency
plt.xlim(0, 0.25)  # Adjust this range as per your expected data range

# Labels, title, and grid
plt.xlabel('Scores', fontsize=12)
plt.ylabel('Retrieval Methods', fontsize=12)
plt.title('Rouge Scores for Normal Queries', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Ensure consistent layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
files_names_req = ["ZeroShot_response.json", "ZeroShot_response_modified.json",
                   "WikiAgent_response_modified.json", "WikiAgent_response.json",
                   "LlamaAgent_response.json", "LlamaAgent_response_modified.json",
                   "tf_idf_bm25_open_1_2_top_100_combined_response.json",
                   "tf_idf_bm25_open_1_1_top_100_combined_response.json",
                   "tf_idf_bm25_open_1_0_top_100_combined_response.json",
                    "tf_idf_bm25_open_1_2_top_100_combined_modified_response.json", "tf_idf_bm25_open_1_1_top_100_combined_modified_response.json", "tf_idf_bm25_open_1_0_top_100_combined_modified_response.json",
                    "tf_idf_bm25_open_1_2_top_100_combined_both_response.json",
                    "tf_idf_bm25_open_1_1_top_100_combined_both_response.json",
                    "tf_idf_bm25_open_1_0_top_100_combined_both_response.json",
                    "reranked_best_answers_1_0.json",
                    "reranked_best_answers_1_1.json",
                    "reranked_best_answers_1_2.json",]

score = {}

for file_name in files_names_req:
    for rouge_score in rouge_scores:
        if rouge_score['file_name'] == file_name:
            score[file_name.replace("_top_100", "").replace("_response","").replace("_combined","")] = rouge_score['average_rouge_l']
```

```python
# Extract keys and values for plotting
labels = list(score.keys())
values = list(score.values())

# Create the plot
plt.figure(figsize=(12, 6))
plt.barh(labels, values, color='skyblue', edgecolor='black')

# Set fixed axis limits for consistency
plt.xlim(0, 0.25)  # Adjust this range as per your expected data range

# Labels, title, and grid
plt.xlabel('Scores', fontsize=12)
plt.ylabel('Retrieval Methods', fontsize=12)
plt.title('Rouge Scores', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Ensure consistent layout
plt.tight_layout()

# Show the plot
plt.show()
```
